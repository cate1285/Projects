```
import pandas as pd

try:
  lista_compra = pd.read_csv("Ventas.csv",sep=";",encoding="utf-8")
except:
  print("El archivo no se encontro")
else:
  lista_compra.rename(columns={"id_customer":"consumidor","id_product":"codigo_producto","nombre":"producto","cantidad":"Cantidad","Ciudad":"ciudad"},inplace=True)
  lista_compra=lista_compra[["consumidor","codigo_producto","producto","Cantidad","ciudad","valor"]]
  lista_compra=lista_compra.dropna()
lista_compra
```

```
try:
  lista_customers=pd.read_csv("Customers.csv",sep=",",encoding="utf-8")
except URLerror:
  print("El archivo no se encontro")
else:
  lista_customers.rename(columns={"Id_Customer":"consumidor","nombre":"primer_nombre","apellido":"Apellido","Ciudad":"ciudad"},inplace=True)
  lista_customers=lista_customers[["consumidor","primer_nombre","Apellido","ciudad"]]
lista_customers
```

```
import pandas as pd 
merged_df = pd.merge(lista_customers, lista_compra, how="inner",on="consumidor") 
merged_df
```

```
def compra_usuario (merged_df):
  return lista_compra.producto.value_counts().sort_values(ascending=True)
print(compra_usuario(lista_compra))
```

```
merged_df=merged_df[["primer_nombre","Cantidad"]]
merged_df
```

```
df2 = merged_df.groupby(['primer_nombre']).agg({'Cantidad': 'sum', })
df2
```

```
import matplotlib.pyplot as plt

def x(lista_compra,productos):
  fig, ax = plt.subplots()
  lista_compra[lista_compra.producto.isin(productos)].consumidor.value_counts(normalize=False).plot(kind="bar")
  ax.set_title("Diagrama tipo Pie con porcenjes de tipos de productos segun consumidores "+ ",".join(productos),loc="center",fontdict = {"fontsize":14,"fontweight":"bold","color":"blue"})
  ax.set_ylabel("")
  plt.show()
  return

x(lista_compra,["Pur Value"])

```
